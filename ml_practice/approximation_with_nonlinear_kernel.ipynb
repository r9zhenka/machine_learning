```python
import numpy as np
from sklearn import svm
```

```python
def func(x):
    return np.sin(0.5 * x) + 0.2 * np.cos(2*x) - 0.1 * np.sin(4 * x) + 3

coord_x = np.expand_dims(np.arange(-4.0, 6.0, 0.1), axis=1)
coord_y = func(coord_x).ravel()
```

```python
svr = svm.SVR(kernel='rbf')
x_train, y_train = coord_x[::3], coord_y[::3]
svr.fit(x_train, y_train)
predict = svr.predict(coord_x)
Q = np.mean([(predict[i] - coord_y[i]) ** 2 for i in range(len(predict))])
```

```python
import matplotlib.pyplot as plt

plt.plot(coord_x, coord_y)
plt.plot(coord_x, predict)
```
