```python
import numpy as np


def func(x):
    return 0.5 * x + 0.2 * x ** 2 - 0.05 * x ** 3 + 0.2 * np.sin(4 * x) - 3


coord_x = np.arange(-4, 6, 0.1)
coord_y = func(coord_x)
K = 10
X = np.array([[xx**i for i in range(K)] for xx in coord_x]) 
Y = coord_y

X_train = X[::2]
Y_train = Y[::2]

F = X_train.T @ X_train / X_train.shape[0]
L, W = np.linalg.eig(F)

WW = sorted(zip(L, W), key=lambda lx: lx[0], reverse=0)
WW = np.array([w[1] for w in WW])
# X_train.shape, WW.shape
G = X @ WW.T

# исследуем зависимость качества аппроксимации от количества признаков
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
Q = []
fig, axes = plt.subplots(5, 2, figsize=(10, 10))
for x in range(1, G.shape[1] + 1):
    G_plot = G[:, :x] # усекаем до x признаков shape = (l, x)
    X_plot = G_plot[::2]  # новое признаковое пространство, прореженное до размера Y_train
    w_plot = np.linalg.inv(X_plot.T @ X_plot) @ X_plot.T @ Y_train # shape = (x,)
    predict = G_plot @ w_plot
    Q.append(mean_squared_error(predict, Y))
    axes.flat[x-1].plot(coord_x, Y)
    axes.flat[x-1].plot(coord_x, predict)
    axes.flat[x-1].set_title(label=f'усечение до {x} из {G.shape[1]} признаков', fontsize=10)
plt.show()
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), Q)
plt.title('MSE от усечения до х признаков')
plt.ylabel('MSE')
plt.xlabel('number of features')
    
```
