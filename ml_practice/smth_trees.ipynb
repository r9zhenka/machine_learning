```python
# probability
p = lambda i, X: np.sum(X == i) / len(X)

# impurity by Gini
Si = lambda X: 1 - sum([p(i, X) ** 2 for i in np.unique(X)])

# information gain
inf_gain = lambda X, yl, yr:  Si(X) - (len(yl) * Si(yl) + len(yr) * Si(yr)) / len(X)
IG_list = []
IG = (0,)
for j in range(len(x_train[0])):
    range_t = np.arange(min(x_train[:, j])+0.1, max(x_train[:, j])-0.1, 0.1)
    for x in range_t:
        tl = np.where(x_train[:, j] < x)[0]
        y_left = y_train[tl]
        y_right = y_train[np.where(x_train[:, j] >= x)[0]]
        IG_list.append((inf_gain(y_train, y_left, y_right), x, j))
    IG = max(max(IG_list, key=lambda x: x[0]), IG, key=lambda x: x[0])

IG, th, fj = IG
```
